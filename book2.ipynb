```python
import pandas as pd
data = pd.read_csv('csr_2021-22_covidfilteredbook1.csv')
data
```

```python
companies = data['Company'].unique()
print(companies)
print(len(companies))
```

```python
total_expenditures = []
covid_expenditures = []
for i in range(len(companies)):
    sum_te = 0
    
    df1 = data.loc[data.Company == companies[i]]
    lst = df1['Actual Amount Spent (Rs.lacs)'].values.tolist()
    for j in range(len(lst)): 
        sum_te += lst[j]
    total_expenditures.append(sum_te)

for k in range(len(companies)):
    wst = []
    df2 = data.loc[data.Covid_Category == 'Covid']
    df3 = df2.loc[data.Company == companies[k]]
    wst = df3['Actual Amount Spent (Rs.lacs)'].values.tolist()
    covid_expenditures.append(sum(wst))
    
```

```python
print(len(total_expenditures))
print(len(covid_expenditures))
print(len(companies))
```

```python
csr_exp_ratio = {}
for l in range(len(companies)):
    if total_expenditures[l] != 0:
        csr_exp_ratio[companies[l]] = covid_expenditures[l]/total_expenditures[l]
    else:
        csr_exp_ratio[companies[l]] = 0
print(csr_exp_ratio)
```

```python
import matplotlib.pyplot as plt
ys = csr_exp_ratio.values()
xs = csr_exp_ratio.keys()

plt.plot(xs, ys)
plt.show()
```
